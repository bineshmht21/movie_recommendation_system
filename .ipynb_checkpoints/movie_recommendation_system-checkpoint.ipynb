```python
import pandas as pd

df = pd.read_csv(r"C:\Users\BineshMahato\.cache\kagglehub\datasets\rahulvyasm\netflix-movies-and-tv-shows\versions\1\netflix_titles.csv",  encoding='ISO-8859-1')

# Define the function to combine selected features
def combine_features(row):
    return ' '.join([
        str(row['listed_in']) if pd.notna(row['listed_in']) else "",  # Assuming this contains genres
        str(row['description']) if pd.notna(row['description']) else "",  # Assuming you want to include description
        str(row['cast']) if pd.notna(row['cast']) else "",  # Handle missing cast data
        str(row['director']) if pd.notna(row['director']) else "",  # Handle missing director data
        str(row['rating']) if pd.notna(row['rating']) else ""  # Optionally, include rating
    ]).strip()

# Apply the function to each row and create the 'combined_features' column
df['combined_features'] = df.apply(lambda row: combine_features(row), axis=1)

# Display the first few rows of the updated dataframe
print(df[['title', 'combined_features']].head())
```

```python
# Convert text data to numerical using CountVectorizer
vectorizer = CountVectorizer()
feature_matrix = vectorizer.fit_transform(df['combined_features'])
```

```python
# Calculate similarity scores
cosine_sim = cosine_similarity(feature_matrix)
```

```python
# Create a function to recommend movies based on cosine similarity
def recommend_movie(movie_title, num_recommendations=5):
    # Find the index of the movie in the dataset
    movie_index = df[df['title'] == movie_title].index[0]
    
    # Get the similarity scores for the given movie
    similar_movies = list(enumerate(cosine_sim[movie_index]))
    
    # Sort movies by similarity score
    sorted_movies = sorted(similar_movies, key=lambda x: x[1], reverse=True)
    
    # Get the titles of the top 'num_recommendations' movies
    recommended_movies = [df['title'][i[0]] for i in sorted_movies[1:num_recommendations+1]]
    
    return recommended_movies
```

```python
# Test the function
print(recommend_movie("The Matrix"))
```
